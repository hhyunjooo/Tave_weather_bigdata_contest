## 공적분

```python
import statsmodels
from statsmodels.tsa.stattools import coint
```

```python
buy_sns_ha_ci = buy_sns_ha[buy_sns_ha['sm_cat'] == '카페트매트']
```

```python
buy_sns_ha_ci
```

```python
buy_sns_ha_ci.plot(x = 'date')   # 날짜를 기준으로 SNS 언급량과 구매량 시각화
```

```python
# 공적분을 통해 SNS언급량과 구매량간의 추세 비교
score, p_value, _ = coint(buy_sns_ha_ci.qty, buy_sns_ha_ci.cnt)
print('Cointegration test p-value: ' ,p_value)
```

```python
# 머신러닝 모델에서 좋은 성능을 보인 상품들에 대해 공적분 분석 실시
# P-value가 0.05 미만으로 관계가 유의한 경우만 [cg_p_val_df]데이터 프레임에 저장

cg_p_val_df = pd.DataFrame(columns=['CG','P-value'])
cg_p_val_df_all = pd.DataFrame(columns=['CG','P-value'])

for cg_ in ['벽걸이형 선풍기','탁상/USB 선풍기','에어컨 리모컨','냉풍기','이동형 에어컨','업소용 선풍기',
            '돈풍기','라디에이터','컨벡터','황토매트','히터','제습기','초음파식 가습기','자연식 가습기',
            '복합식 가습기','가열식 가습기','에어워셔','공기청정기','공기정화 용품']:
    buy_sns_ha_ci = buy_sns_ha[buy_sns_ha['sm_cat'] == cg_]
    score, p_value, _ = coint(buy_sns_ha_ci.qty, buy_sns_ha_ci.cnt)
    print('Cointegration test p-value: ', cg_ ,p_value)
    cg_p_val_df_all = cg_p_val_df_all.append({'CG':cg_, 'P-value':p_value}, ignore_index=True)
    if p_value <=0.05:
        cg_p_val_df = cg_p_val_df.append({'CG':cg_, 'P-value':p_value}, ignore_index=True)
    
```

```python
cg_p_val_df_all.to_csv('../활용 데이터셋/cg_p_val_df_all.csv', encoding='cp949')
cg_p_val_df.to_csv('../활용 데이터셋/cg_p_val_df_all.csv', encoding='cp949')
```

```python
# 필요한 라이브러리 불러오기
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.dates import DateFormatter
from matplotlib import dates as mpld
plt.rc('font', size=30) 

from statsmodels.tsa.seasonal import seasonal_decompose

from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
items_list =['제습기','돈풍기','공기정화 용품','공기청정기']

items_dict = {'제습기':'제습기','돈풍기':'돈풍기','공기정화 용품':'공기정화_용품','공기청정기':'공기청정기'}

for i in items_list:
    print(i,':',items_dict[i])
```

```python
for item in items_list:
#     sns 언급데이터 이상탐지
    buy_sns_ha_cnt = buy_sns_ha[buy_sns_ha['sm_cat'] == item]
    buy_sns_ha_cnt.drop(['sm_cat', 'qty'], axis=1, inplace = True)
    buy_sns_ha_cnt.set_index('date', inplace=True)
    
    temp_cnt = buy_sns_ha_cnt
    
    # 앞의 70% 데이터만 학습에 사용
    train_cnt = temp_cnt.iloc[:int(len(temp_cnt)*0.7)]
    test_cnt = temp_cnt.iloc[int(len(temp_cnt)*0.7):]
    
    seasonal_decompose(temp_cnt['cnt'],model='additive',period=12)
    
    
    # ES 모델을 만들어 학습하고 전체 데이터에 대해 예측한다.
    # additive는 경향성이 일정함을 의미하고, 경향성 변동폭이 있을 때는 multiplicative를 사용
    model = ExponentialSmoothing(train_cnt, trend='additive', seasonal='additive').fit()
    
    prediction = model.predict(start=temp_cnt.index[0], end=temp_cnt.index[-1])

    """Brutlag Algorithm"""
    PERIOD = 12        # The given time series has seasonal_period=12
    GAMMA = 0.1  # the seasonility component
    SF = 1.96          # brutlag scaling factor for the confidence bands.
    UB = []            # upper bound or upper confidence band
    LB = []            # lower bound or lower confidence band

    # 실측치와 예측치를 비교하는 자료구조
    difference_array = []
    dt = []
    difference_table = {"actual": temp_cnt, "predicted": prediction, "difference": difference_array, "UB": UB, "LB": LB}

    # brutlag 알고리즘
    # 12개월 이전의 실측/결측 차이에 0.63, 이번달 차이에 0.37 정도의 가중치를 주어 저장
    for i in range(len(prediction)):
        diff = temp_cnt.iloc[i]-prediction.iloc[i]
        if i < PERIOD:
            dt.append(GAMMA*abs(diff))
        else:
            dt.append(GAMMA*abs(diff) + (1-GAMMA)*dt[i-PERIOD])

    # 저장된 실측/결측 차이를 예측치의 95% 신뢰구간(1.96)으로 반영하여 Upper/Lower Band 계산
        difference_array.append(diff)
        UB.append(prediction[i]+SF*dt[i])
        LB.append(prediction[i]-SF*dt[i])
    
    """Classification of data points as either normal or anomaly"""
    normal_cnt = []
    normal_cnt_date = []
    anomaly_cnt = []
    anomaly_cnt_date = []

    # 신뢰구간을 벗어나는지 판단하여 normal, anomaly 결정
    for i in range(len(temp_cnt.index)):
        if (UB[i] <= temp_cnt.iloc[i]).bool() and i > PERIOD:
            anomaly_cnt_date.append(temp_cnt.index[i])
            anomaly_cnt.append(temp_cnt.iloc[i][0])
        else:
            normal_cnt_date.append(temp_cnt.index[i])
            normal_cnt.append(temp_cnt.iloc[i][0])
        
    globals()['anomaly_cnt_{}'.format(items_dict[item])]= pd.DataFrame({"date": anomaly_cnt_date, "value": anomaly_cnt})
    globals()['anomaly_cnt_{}'.format(items_dict[item])].set_index('date', inplace=True)
    globals()['nomaly_cnt_{}'.format(items_dict[item])] = pd.DataFrame({"date": normal_cnt_date, "value": normal_cnt})
    globals()['nomaly_cnt_{}'.format(items_dict[item])].set_index('date', inplace=True)

    # plotting
    plt.figure(figsize=(24,12))
    plt.title(item+str('_소셜 데이터'))
    plt.plot(globals()['nomaly_cnt_{}'.format(items_dict[item])].index, globals()['nomaly_cnt_{}'.format(items_dict[item])], 'o', color='green')
    plt.plot(globals()['anomaly_cnt_{}'.format(items_dict[item])].index, globals()['anomaly_cnt_{}'.format(items_dict[item])], 'o', color='red')
    plt.plot(temp_cnt.index, UB, linestyle='--', color='grey')
    plt.plot(temp_cnt.index, LB, linestyle='--', color='grey')
    plt.legend(['Normal', 'Anomaly', 'Upper Bound', 'Lower Bound'])
    plt.show()
    
# ========================================================================================================================    
#     구매 데이터 이상탐지
    buy_sns_ha_qty = buy_sns_ha[buy_sns_ha['sm_cat'] == item]
    buy_sns_ha_qty.drop(['sm_cat', 'cnt'], axis=1, inplace = True)
    buy_sns_ha_qty.set_index('date', inplace=True)
    temp_qty = buy_sns_ha_qty

    train_qty = temp_qty.iloc[:int(len(temp_qty)*0.7)]
    test_qty = temp_qty.iloc[int(len(temp_qty)*0.7):]

    seasonal_decompose(temp_qty['qty'],model='additive',period=12)

    model = ExponentialSmoothing(train_qty, trend='additive', seasonal='additive').fit()
    
    prediction = model.predict(start=temp_qty.index[0], end=temp_qty.index[-1])
    
    """Brutlag Algorithm"""
    PERIOD = 12        # The given time series has seasonal_period=12
    GAMMA = 0.1  # the seasonility component
    SF = 1.96          # brutlag scaling factor for the confidence bands.
    UB = []            # upper bound or upper confidence band
    LB = []            # lower bound or lower confidence band
    
    # 실측치와 예측치를 비교하는 자료구조
    difference_array = []
    dt = []
    difference_table = {"actual": temp_qty, "predicted": prediction, "difference": difference_array, "UB": UB, "LB": LB}

    # brutlag 알고리즘
    # 12개월 이전의 실측/결측 차이에 0.63, 이번달 차이에 0.37 정도의 가중치를 주어 저장
    for i in range(len(prediction)):
        diff = temp_qty.iloc[i]-prediction.iloc[i]
        if i < PERIOD:
            dt.append(GAMMA*abs(diff))
        else:
            dt.append(GAMMA*abs(diff) + (1-GAMMA)*dt[i-PERIOD])

    # 저장된 실측/결측 차이를 예측치의 95% 신뢰구간(1.96)으로 반영하여 Upper/Lower Band 계산
        difference_array.append(diff)
        UB.append(prediction[i]+SF*dt[i])
        LB.append(prediction[i]-SF*dt[i])
    
    """Classification of data points as either normal or anomaly"""
    normal_qty = []
    normal_qty_date = []
    anomaly_qty = []
    anomaly_qty_date = []

# 신뢰구간을 벗어나는지 판단하여 normal, anomaly 결정
    for i in range(len(temp_qty.index)):
        if (UB[i] <= temp_qty.iloc[i]).bool() and i > PERIOD:
            anomaly_qty_date.append(temp_qty.index[i])
            anomaly_qty.append(temp_qty.iloc[i][0])
        else:
            normal_qty_date.append(temp_qty.index[i])
            normal_qty.append(temp_qty.iloc[i][0])
        
    globals()['anomaly_qty_{}'.format(items_dict[item])]= pd.DataFrame({"date": anomaly_qty_date, "value": anomaly_qty})
    globals()['anomaly_qty_{}'.format(items_dict[item])].set_index('date', inplace=True)
    globals()['nomaly_qty_{}'.format(items_dict[item])] = pd.DataFrame({"date": normal_qty_date, "value": normal_qty})
    globals()['nomaly_qty_{}'.format(items_dict[item])].set_index('date', inplace=True)

    # plotting
    plt.figure(figsize=(24,12))
    plt.title(item+str('_판매 데이터'))
    plt.plot(globals()['nomaly_qty_{}'.format(items_dict[item])].index, globals()['nomaly_qty_{}'.format(items_dict[item])], 'o', color='green')
    plt.plot(globals()['anomaly_qty_{}'.format(items_dict[item])].index, globals()['anomaly_qty_{}'.format(items_dict[item])], 'o', color='red')
    plt.plot(temp_qty.index, UB, linestyle='--', color='grey')
    plt.legend(['Normal', 'Anomaly', 'Upper Bound', 'Lower Bound'])
    plt.show()
```

```python
# 구매데이터의 이상치 중 SNS데이터의 이상치와 일치하는 비율은 qty_percentage_df에 저장
# SNS데이터의 이상치 중 구매데이터의 이상치와 일치하는 비율은 cnt_percentage_df에 저장

qty_percentage_df = pd.DataFrame(columns=['item','percent(%)'])
cnt_percentage_df = pd.DataFrame(columns=['item','percent(%)'])

```

```python
items =['제습기','돈풍기','공기정화_용품','공기청정기']
qty_date_dict={}
cnt_date_dict={}


for item in items:
    globals()['date_list_cnt_{}'.format(item)] = []
    globals()['date_list_qty_{}'.format(item)] = []

    for cnt_date in globals()['anomaly_cnt_{}'.format(item)].index:
        for qty_date in globals()['anomaly_qty_{}'.format(item)].index:
            # 날짜 파라미터 설정(7일 이내)
            # SNS 데이터 이상치 탐지일로부터 7일이네에 있는 구매 데이터 이상치 탐지일 선별
            # 상품별로 이상치 정리
            if  (qty_date - cnt_date <= timedelta(days=7)) & (qty_date - cnt_date>= timedelta(days=0)):
                globals()['date_list_cnt_{}'.format(item)].append(cnt_date)     # 해당 구매 데이터 이상치 날짜
                globals()['date_list_qty_{}'.format(item)].append(qty_date)     # 해당 SNS 데이터 이상치 날짜
    
    
    # 구매데이터 이상치 탐지일 날짜순으로 정렬
    date_list_qty_set = set(globals()['date_list_qty_{}'.format(item)])       
    globals()['date_list_qty_{}'.format(item)] = list(date_list_qty_set)       
    globals()['date_list_qty_{}'.format(item)].sort()
    
    # 구매데이터의 이상치 중 SNS데이터의 이상치와 일치하는 비율 계산
    percent = len(globals()['date_list_qty_{}'.format(item)])/len(globals()['anomaly_qty_{}'.format(item)]) * 100
    qty_date_dict[item]=percent
    qty_percentage_df = qty_percentage_df.append({'item':str(item), 'percent(%)':percent}, ignore_index=True)
    
    
    
    # SNS 데이터 이상치 탐지일 날짜순으로 정렬
    date_list_cnt_set = set(globals()['date_list_cnt_{}'.format(item)])
    globals()['date_list_cnt_{}'.format(item)] = list(date_list_cnt_set)
    globals()['date_list_cnt_{}'.format(item)].sort()
    
    # SNS데이터의 이상치 중 구매데이터의 이상치와 일치하는 비율 계산
    percent = len(globals()['date_list_cnt_{}'.format(item)])/len(globals()['anomaly_cnt_{}'.format(item)]) * 100
    cnt_date_dict[item]=percent
    cnt_percentage_df = cnt_percentage_df.append({'item':item, 'percent(%)':percent}, ignore_index=True)
    
```
