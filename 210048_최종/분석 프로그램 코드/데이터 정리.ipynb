## 종합(불쾌지수/열지수/체감온도)

```python
weather_total = pd.merge(weather_201819, humidity_201819 , how='outer', on='일시')
```

```python
weather_total = pd.merge(weather_total,dust_201819,how='outer', left_on='일시', right_on='일시')
```

```python
weather_total.info()
```

```python
weather_total = weather_total.fillna(weather_total['일 미세먼지 농도(㎍/㎥)'].mean())
```

```python
weather_total.info()
```

### 불쾌지수

```python
weather_total['불쾌지수'] = 1.8 * weather_total['평균기온(°C)'] - 0.55 * (1 - weather_total['평균 상대습도(%)']/100) * (1.8 * weather_total['평균기온(°C)'] - 26) + 32
```

```python
weather_total
```

### 열지수

```python
weather_total['열지수']=0.5 * (1.8*weather_total['평균기온(°C)'] +32 + 61.0 + ((1.8*weather_total['평균기온(°C)']+32-68.0) * 1.2) + (weather_total['평균 상대습도(%)']/100*0.094))
```

```python
weather_total
```

### 체감온도

```python
weather_total['체감온도']=13.12+(0.6215*weather_total['평균기온(°C)'])-(11.37*(weather_total['평균 풍속(m/s)']**(0.16))) + 0.3965*weather_total['평균기온(°C)'] * (weather_total['평균 풍속(m/s)']**0.16)
```

```python
weather_total
```
